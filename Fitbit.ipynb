Details:

- 7-day measurements from 18 participants
- each responded to Ecologic Momemntary Assessment (EMA) surveys (every two hours) sent to them via Twilio text messaging

````
import pandas as pd
fitbit = pd.read_csv('fitbit.csv')
````

````
pd.set_option("display.max_rows", None, "display.max_columns", None)

from IPython.display import Javascript
display(Javascript('''google.colab.output.setIframeHeight(0, true, {maxHeight: 14821})'''))

fitbit.head(61)
````

````
import numpy as np

s=np.unique(fitbit.Subject)
print( s, len(s) )
````

````
q=np.where( fitbit['Subject']== 6 )[0]         
fb_hr = fitbit.iloc[ q, fitbit.columns.get_loc( 'HR' ) ].values
fb_steps = fitbit.iloc[ q, fitbit.columns.get_loc( 'Steps' ) ].values
fb_elevation = fitbit.iloc[ q, fitbit.columns.get_loc( 'Elevation' ) ].values

import matplotlib.pyplot as plt

from IPython.display import Javascript
display(Javascript('''google.colab.output.setIframeHeight(0, true, {maxHeight: 14821})'''))

H=24
W=12
plt.clf();

fig,axes=plt.subplots(2,1, figsize=( H, W ));

a=0
axes[a].plot(fb_steps, label='Steps');
axes[a].plot(fb_hr, label='Heart rate');
axes[a].legend();
axes[a].set_title('Subject %d' % 6 );

a=1
axes[a].plot(fb_elevation, label='Elevation');
axes[a].set_xlabel('Seconds');


import scipy.stats

q=np.where( np.isnan(fb_hr)==False )[0]
q2=np.where( np.isnan(fb_steps)==False)[0]
q=np.intersect1d(q,q2)

st=scipy.stats.spearmanr(fb_hr[q], fb_steps[q])   # Spearman's rho

print( 'Correlation between heart rate and step-count:\n\tcorr=%.3f, pvalue=%.14f' %(st.correlation, st.pvalue ))
print( '\tCalculated from %.2f fraction of data, '% (len(q)/ len(fb_hr)), len(fb_hr),'in total length' )


q=np.where( np.isnan(fb_hr)==False )[0]
q2=np.where( np.isnan(fb_elevation)==False)[0]
q=np.intersect1d(q,q2)
st=scipy.stats.spearmanr(fb_hr[q], fb_elevation[q])   # Spearman's rho

print( 'Correlation between heart rate and elevation:\n\tcorr=%.3f, pvalue=%.14f' %(st.correlation, st.pvalue ))
print( '\tCalculated from %.2f fraction of data, '% (len(q)/ len(fb_hr)), len(fb_hr),'in total length\n\n' )

````

- To demonstrate plotting subplots arranged in a grid manner

```
import umap
import matplotlib.pyplot as plt
import numpy as np
import random

fig = plt.figure(figsize=( 10, 5 ))  

nps2= 4

axes=fig.subplots(2,nps2)  # rows by cols

a=b=-1
if ( 'embeddings' in globals() )==False:
  embeddings = dict()


for d in range(8):

  data = np.random.randint(0, d+1,  (2000, 90 )) # scale of d changes

  #data = AFdataTmNormalized_imputed[mxDaysTillAbl-90:mxDaysTillAbl,d,:].transpose()        
  data[np.isnan(data)]= 0

  if 0:    
    u_map = umap.UMAP( n_components = 2 )                 
    u_map.fit( data )                
    embeddings[ d ] = u_map.transform( data ) 

  dat = embeddings[ d ]
  b+=1       
  if (b % ( nps2 ) ) == 0:
      a+=1; b=0   
  axes[a,b].plot(dat[:,0], dat[:,1], '.' ) 
  axes[a,b].set_title( d )

axes[a,b].legend(markerscale=6)# scale up size of the markers in legend
       
fig.tight_layout()         

#fig.savefig('test.png')        

```
